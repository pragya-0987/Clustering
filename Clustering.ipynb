## **THEORETICAL QUESTIONS**

**1.  What is unsupervised learning in the context of machine learning ?**

Unsupervised learning is a type of machine learning where the model learns patterns and structures from data **without labeled outputs**. Instead of being told what the “right answer” is (like in supervised learning), the algorithm tries to find hidden patterns, groupings, or features on its own—commonly used for **clustering**, **dimensionality reduction**, and **anomaly detection**.

**2. How does K-Means clustering algorithm work ?**

K-Means clustering works by dividing data into **K groups (clusters)** based on feature similarity. Here's how it works in short:

1. **Initialize**: Choose K random points as initial centroids.
2. **Assign**: Assign each data point to the nearest centroid (based on distance).
3. **Update**: Recalculate the centroids as the average of all points in each cluster.
4. **Repeat**: Keep assigning and updating until the centroids no longer change much (converge).

The result: similar data points are grouped together in the same cluster.

**3.  Explain the concept of a dendrogram in hierarchical clustering.**

A **dendrogram** is a tree-like diagram used in **hierarchical clustering** to show how data points are grouped together step by step.

- Each **leaf** represents a single data point.
- As you move up the tree, branches **merge clusters** based on their similarity.
- The **height** of each merge shows the **distance** (or dissimilarity) between clusters.

You can **cut the dendrogram** at a certain level to decide how many clusters you want.

**4. What is the main difference between K-Means and Hierarchical Clustering ?**

The main difference is:

- **K-Means** requires you to **predefine the number of clusters (K)** and is **faster** for large datasets.
- **Hierarchical clustering** **doesn't need K upfront** and builds a **tree (dendrogram)** of clusters, but it’s **slower** and more **computationally intensive**.

In short:  
**K-Means = flat, fast, needs K**  
**Hierarchical = tree-like, flexible, slower**

**5. What are the advantages of DBSCAN over K-Means ?**

**DBSCAN** has several advantages over **K-Means**:

1. **No need to specify number of clusters** beforehand.
2. Can find **arbitrarily shaped clusters**, not just spherical ones.
3. **Handles noise/outliers** well by labeling them separately.
4. Better for **clusters with varying densities**.

In short:  
**DBSCAN = more flexible, handles noise, no K needed**  
**K-Means = simpler, but limited to spherical clusters & needs K**

**6. When would you use Silhouette Score in clustering ?**

You’d use the **Silhouette Score** to **evaluate how well your clustering performed**.

It measures how similar each point is to its **own cluster** vs. other clusters:

- Score ranges from **-1 to 1**
  - **Close to 1** = well clustered  
  - **Close to 0** = overlapping clusters  
  - **Negative** = wrong cluster


Use **Silhouette Score** to check **clustering quality** and help **choose the best number of clusters**.

**7. What are the limitations of Hierarchical Clustering ?**

Limitations of **Hierarchical Clustering** in short:

1. **Not scalable** for large datasets (high time & space complexity).
2. **Irreversible merges** – once clusters are joined, can’t undo.
3. Sensitive to **noise and outliers**.
4. Choosing the **right number of clusters** from the dendrogram can be subjective.

In short:  
**Slow, inflexible, sensitive to noise**, and **hard to scale**.

**8.  Why is feature scaling important in clustering algorithms like K-Means ?**

Feature scaling is important in **K-Means** because:

- K-Means uses **distance (like Euclidean)** to group points.
- If features are on different scales, **larger-scale features dominate** the distance.
- This can lead to **biased or incorrect clusters**.

**Scaling ensures all features contribute equally** to clustering.

**9.  How does DBSCAN identify noise points ?**

**DBSCAN** identifies **noise points** as those that:

- Have **fewer neighbors** than a set threshold (**MinPts**)  
- Within a given **radius** (**ε/epsilon**)

If a point isn't part of any dense region (cluster), it’s labeled as **noise (or outlier)**.


**DBSCAN marks points as noise** if they’re **too isolated** to belong to any cluster.

**10.  Define inertia in the context of K-Means.**

In **K-Means**, **inertia** is the **sum of squared distances** between each data point and its **assigned cluster centroid**.

It measures how **compact** the clusters are — lower inertia means **tighter clusters**.

In short:  
**Inertia = measure of cluster tightness** — lower is better.

**11.  What is the elbow method in K-Means clustering ?**


The **Elbow Method** in **K-Means clustering** helps determine the optimal number of clusters (K).

- It involves plotting **inertia** (sum of squared distances) for different values of K.
- The "elbow" point, where the inertia starts to **decrease more slowly**, indicates the ideal K.


The **elbow** represents the point where adding more clusters no longer significantly improves the model.

**12.  Describe the concept of "density" in DBSCAN.**

In **DBSCAN**, **density** refers to the number of data points within a specific **radius (ε)** around a point.

- A region is considered **dense** if it has at least **MinPts** (minimum points) within this radius.
- **High-density areas** form clusters, while **low-density areas** are considered **noise**.

In short:  
**Density** in DBSCAN is about how many points are packed together within a certain distance, helping to define clusters.

**13.  Can hierarchical clustering be used on categorical data?**

Yes, **hierarchical clustering** can be used on **categorical data**, but it requires a **suitable distance metric** for categorical features, such as:

- **Jaccard distance** or **Hamming distance**.

These metrics measure similarity or dissimilarity between categorical variables, allowing hierarchical clustering to group categorical data.


Hierarchical clustering can work on categorical data if an appropriate **distance measure** is used.

**14. What does a negative Silhouette Score indicate ?**

A **negative Silhouette Score** indicates that the data point is likely **assigned to the wrong cluster**. It suggests that the point is closer to a different cluster than its own, meaning the clustering structure may need improvement.

In short:  
A **negative score** means poor cluster assignment and potential overlap between clusters.

**15.  Explain the term "linkage criteria" in hierarchical clustering.**

**Linkage criteria** in **hierarchical clustering** determine how the **distance between clusters** is calculated during the merging process. Common linkage methods include:

1. **Single linkage**: Distance between the closest points of two clusters.
2. **Complete linkage**: Distance between the farthest points of two clusters.
3. **Average linkage**: Average distance between all pairs of points in two clusters.
4. **Ward's linkage**: Minimizes the increase in variance within clusters.


**Linkage criteria** control how clusters are merged based on the chosen distance metric.

**16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?**

**K-Means** can perform poorly on data with varying **cluster sizes** or **densities** because:

- It assumes **spherical, equally sized clusters**.
- It uses **Euclidean distance**, which doesn’t handle clusters of different shapes or densities well.
- **Outliers** or sparse areas can distort the centroids, leading to incorrect cluster assignments.

In short:  
K-Means struggles with **non-spherical clusters** and **varying densities**.

**17. What are the core parameters in DBSCAN, and how do they influence clustering ?**

The core parameters in **DBSCAN** are:

1. **ε (epsilon)**: The **radius** that defines the neighborhood around a point. Smaller values lead to more, smaller clusters; larger values may merge clusters.
2. **MinPts**: The minimum number of points required to form a dense region (a cluster). Higher values make DBSCAN more selective in forming clusters.


**ε** controls **neighborhood size**, and **MinPts** controls **density requirement** for clusters.

**18.  How does K-Means++ improve upon standard K-Means initialization ?**

**K-Means++** improves upon standard **K-Means** initialization by selecting initial centroids more intelligently. It chooses the first centroid randomly, then selects subsequent centroids with a higher probability based on their distance from existing centroids. This leads to better cluster quality and faster convergence compared to random initialization.

In short:  
**K-Means++** spreads out initial centroids, improving clustering results and convergence speed.

**19. What is agglomerative clustering ?**

**Agglomerative clustering** is a type of **hierarchical clustering** where each data point starts as its own cluster, and clusters are progressively merged based on similarity. The merging continues until all points are in a single cluster or a stopping criterion is met.



**20.  What makes Silhouette Score a better metric than just inertia for model evaluation?**

**Silhouette Score** is often a better metric than **inertia** because it not only measures **compactness** (like inertia) but also considers **separation** between clusters. A high Silhouette Score indicates both **tight clusters** and **clear separation** between them, while inertia can decrease with more clusters even if they overlap or are poorly separated.

In short:  
**Silhouette Score** evaluates both **cluster cohesion** and **separation**, while inertia only measures compactness.

**21.  Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

**29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data with 300 samples and 3 centers
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# List to store silhouette scores
silhouette_scores = []

# Evaluate silhouette scores for k = 2 to 5
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

# Plot the silhouette scores as a bar chart
plt.bar(range(2, 6), silhouette_scores, color='skyblue')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K = 2 to 5")
plt.xticks(range(2, 6))
plt.show()
```

**31.  Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=300, centers=2, cluster_std=1.5, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Set up the grid for plotting decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict the cluster for each point in the grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.contourf(xx, yy, Z, alpha=0.3, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', edgecolors='k', s=50)
plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

**43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate synthetic noisy circular data
X, _ = make_circles(n_samples=500, noise=0.1, factor=0.5, random_state=42)

# Standardize the data (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X_scaled)

# Plot the clustering results side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# K-Means clustering plot
axes[0].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, alpha=0.6)
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                 s=200, c='red', marker='X', label='Cluster Centers')
axes[0].set_title("K-Means Clustering")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")
axes[0].legend()

# DBSCAN clustering plot
axes[1].scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', s=50, alpha=0.6)
axes[1].set_title("DBSCAN Clustering")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

# Show the plot
plt.tight_layout()
plt.show()
```

**44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.**

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Calculate the silhouette scores for each sample
silhouette_vals = silhouette_samples(X_scaled, y_kmeans)

# Plot the silhouette coefficients for each sample
plt.figure(figsize=(8, 6))
plt.bar(range(len(silhouette_vals)), silhouette_vals, color='skyblue')
plt.axhline(y=0, color='gray', linestyle='--')
plt.title("Silhouette Coefficients for Each Sample in K-Means Clustering")
plt.xlabel("Sample Index")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

**47.  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.**

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate synthetic data with 3 centers (blobs), add noise
X, _ = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.0)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Count the number of clusters and noise points
# Noise points are labeled as -1 in DBSCAN
clusters = len(set(y_dbscan)) - (1 if -1 in y_dbscan else 0)
noise_points = list(y_dbscan).count(-1)

# Print the counts
print(f"Number of clusters: {clusters}")
print(f"Number of noise points: {noise_points}")
```
